### VIX - Volatility Index
##### Higher VIX correlates with higher market turbulence and uncertainity; whereas a lower VIX indicates stability.

```python
start_date = '1990-01-01'
end_date =  '2024-10-15'
```

```python
vix_df = pdr.DataReader('VIXCLS', 'fred', start_date, end_date)
vix_df.head()
```

```python
count_nan = vix_df['VIXCLS'].isna().sum()
print("Number of NaN values is",count_nan)
```

```python
vix_df = vix_df.dropna()
```

```python
vix_mean = vix_df['VIXCLS'].mean()
print('Mean of the dataset:', round(vix_mean,5))
```

```python
plt.plot(vix_df[-250:], color='black', linewidth=1.5, label='Change in VIX')
plt.axhline(y=vix_mean, color='red', linestyle='dashed')
plt.grid()
plt.legend()
plt.show()
```

```python
vix_df = vix_df.diff(periods=1, axis=0)
vix_df.head()
```

```python
vix_df = vix_df.iloc[1:,:]
vix_df.head()
```

```python
vix_mean = vix_df['VIXCLS'].mean()
print('Mean of the dataset:', round(vix_mean,5))
```

```python
plt.plot(vix_df[-250:], color='black', linewidth=1.5, label='Change in VIX')
plt.axhline(y=vix_mean, color='red', linestyle='dashed')
plt.grid()
plt.legend()
plt.show()
```
